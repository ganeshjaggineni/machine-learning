```python
#IMPORTING THE REQUIRED LIBRARIES
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import plot_tree
```

```python
#LOADING THE DATASET IRIS
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
#SPLIT THE DATASET INTO TRAINING AND TESTING
```

```python
x_train, x_test, y_train, y_test = train_test_split = train_test_split(x,y,test_size=0.3, random_state = 42)
```

```python
#CREATE RANDOM FOREST CLASSIFIER
```

```python
rf_classifier = RandomForestClassifier(n_estimators=3, random_state=42, oob_score=True)#oob = out of bag
```

```python
#TRAIN TH CLASSIFIER ON THE TRAINING DATA
```

```python
rf_classifier.fit(x_train, y_train)
```

```python
#MAKE PREDICTIONS ON THE TEST DATA
y_pred = rf_classifier.predict(x_test)
```

```python
#CALCULATING ACCURACY SCORE
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy:{accuracy:.2f}')
```

```python
#CALCULATE AND DISPLAY THE CONFUSION MATRIX
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print("confusion matrix")
print(conf_matrix)
```

```python
#GRAPH VISUALIZATION
```

```python
plt.figure(figsize=(40, 16))
for i, tree in enumerate(rf_classifier.estimators_):
    plt.subplot(5, 5, i + 1)
    plot_tree(tree, filled=True, feature_names=iris.feature_names, class_names=iris.target_names, fontsize=10)
    plt.title(f'Tree {i + 1}')
plt.tight_layout()
plt.show()
```
