```python
#IMPORTING REQUIRED LIBRARIES
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#LOAD THE IRIS DATASET
```

```python
iris = datasets.load_iris()
x = iris.data[: 100, :2] #select two features and two calsses fr
y = iris.target[:100]
```

```python
#SPLIT THE DATA INTO TRAINING AND TESTING SETS
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#INSTANCE OF SVM WITH A LINEAR KERNEL
```

```python
svm_classifier = SVC(kernel='linear',C=1.0)
svm_classifier.fit(x_train,y_train)
```

```python
#MAKE PREDICTIONS ON THE TEST SET
```

```python
y_pred = svm_classifier.predict(x_test)
```

```python
#CALCULATE THE ACCURACY
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ",accuracy)
```

```python
#PLOT THE DATA POINTS
plt.scatter(x[:, 0], x[:, 1],c=y,cmap='coolwarm') #c = clolor and y = class lables and cmap = colormap
#PLOT THE DECISION BOUNDARY
ax = plt.gca()  #ax = axis
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50),
                     np.linspace(ylim[0], ylim[1], 50))
z = svm_classifier.decision_function(np.c_[xx.ravel(),yy.ravel()])
z = z.reshape(xx.shape)
#PLOT THE DECISION BOUNDARY AND SUPPORT VECTORS
plt.contour(xx,yy,z,colors='k',levels=[-1, 0, 1], aplpha = 0.5, linestyles = ['--','-','--'])
plt.scatter(svm_classifier.support_vectors_[:, 0],
            svm_classifier.support_vectors_[:, 1],s = 100,
            linewidth = 1,facecolors='none',edgecolors='k')
plt.title(f'SVM with Linear Kernel\nAccuracy:{accuracy:.2f}')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
