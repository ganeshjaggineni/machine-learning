# IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
```

# READING DATASET

```python
data = pd.read_csv('auto-mpg.csv')
data
```

# HEAD OF DATASET

```python
data.head()
```

# TAIL OF DATASET

```python
data.tail()
```

# SELECTING NUMERIC COLUMNS

```python
numeric_cols = data.select_dtypes(include = np.number)
numeric_cols
```

# FINDING STATISTICAL FUNCTIONS

```python
mean_data = numeric_cols.mean()
median_data = numeric_cols.median()
standard_deviation_data = numeric_cols.std()
variance_data = numeric_cols.var()
```

# CREATING DATASET OF STATISTICAL FUNCTIONS

```python
pd.DataFrame({'mean':mean_data,
            'median':median_data,
            'variance':variance_data,
            'standard deviation':standard_deviation_data})
```
