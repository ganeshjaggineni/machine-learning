# DATA VISUALIZATION : BOX PLOT, SCATTER PLOT, HISTOGRAM

# IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# LOADING THE DATASET

```python
df = pd.read_csv("auto-mpg.csv")
df.head()
```

# BOX PLOT FOR ENTIRE DATASET

```python
plt.figure(figsize = (12, 6))
sns.boxplot(data = df)
plt.title("box plot for entire data set")
plt.show()
```

# BOX PLOT FOR MPG ATTRIBUTE 

```python
plt.figure(figsize=(5,6))
sns.boxplot(data = df['mpg'])
plt.title("BOX PLOT FOR MPG ATTRIBUTE")
plt.show()
```

# QUARTILE RANGES & EXTREME VALUES

```python
q1 = df['mpg'].quantile(0.25)
q2 = df['mpg'].quantile(0.5)
q3 = df['mpg'].quantile(0.75)
min_val = df['mpg'].min()
max_val = df['mpg'].max()
```

```python
plt.figure(figsize = (8,5))
sns.boxplot(data = df['mpg'], orient='v')
plt.title('Box plot for mpg attribute')
plt.ylabel('MPG')
plt.text(0.85,q1,f'Q1:{q1:.2f}',ha = 'left',va='center',fontweight='bold')
plt.text(0.85,q2,f'Q2:{q2:.2f}',ha = 'left',va='center',fontweight=100)
plt.text(0.85,min_val,f'Min:{min_val}',fontweight=100)
```

# HISTOGRAM FOR ENTIRE DATA SET



```python
plt.figure(figsize= (12,8))
df.hist(bins=50)
plt.suptitle("hist for each numerical attribute")
plt.tight_layout()
plt.show()
```

# HISTOGRMA FOR DISPLACEMENT

```python
plt.figure(figsize=(10,8))
sns.histplot(df['displacement'],bins=10)
```

# SCATTER PLOT FOR MPG VS WEIGHT

```python
df.plot.scatter(x='mpg',y='weight',s=30,color='green')
```
