# Data visualization: Box plot,Scatter plot,histogram

#                          step-1 : importing libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#                         LOADING THE DATASET

```python
df=pd.read_csv("auto-mpg.csv")
print(df.head())
```

#                    boxplot for entire dataset

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.title("BOXPLOT FOR OVERALL DATA")
plt.xticks(rotation=45)
plt.show()
```

```python
df.boxplot(figsize=(10,6))
```

#                historgram for entire dataset

```python
plt.figure(figsize=(12,8))
df.hist(bins=50)
plt.suptitle('histograms for each numerical attribute in dataset')
plt.tight_layout()
plt.show()
```

```python
df.hist()
```

#      BOX PLOT FOR MPG ATTRIBUTE

```python
plt.figure(figsize=(5,6))
sns.boxplot(y=df['mpg'])
plt.title('BOXPLOT FOR MPG ATTRIBUTE')
plt.ylabel('mpg')
plt.show()
```

# BOX PLOT FOR CYLINDER ATTRIBUTE

```python
plt.figure(figsize=(5,6))
sns.boxplot(data=df['cylinders'],orient='v')
plt.title('BOXPLOT FOR CYLINDER ATTRIBUTE')
plt.ylabel('cylinder')
plt.show()
```

# BOX PLOT FOR DISPLACEMENT ATTRIBUTE

```python
plt.figure(figsize=(5,6))
sns.boxplot(data=df['displacement'],orient='v')
plt.title('BOXPLOT FOR DISPLACEMENT ATTRIBUTE')
plt.ylabel('displacement')
plt.show()
```

# BOX PLOT FOR MODEL YEAR ATTRIBUTE

```python
plt.figure(figsize=(5,6))
sns.boxplot(data=df['model year'],orient='v')
plt.title('BOXPLOT FOR MODEL YEAR ATTRIBUTE')
plt.ylabel('model year')
plt.show()
```

# QUARTILE RANGES AND EXTREME VALUES

```python
q1 = df['mpg'].quantile(0.25)
q2 = df['mpg'].quantile(0.5)
q3 = df['mpg'].quantile(0.75)
min_value = df['mpg'].min()
max_value = df['mpg'].max()


#create verticlal box plot for the 
plt.figure(figsize=(8,5))
sns.boxplot(data=df['mpg'],orient='V',showfliers=False)
plt.title('BOX PLOT FOR MPG ATTRIBUTE')
plt.ylabel('mpg')

#annotate the quartile ranges and extreme
plt.text(0.85, q1, f'Q1: {q1:.2f}',ha='left',va='center',fontweight='bold')
plt.text(0.85, q2, f'Q2 (median): {q2:.2f}',ha='left',va='center',fontweight='bold')
plt.text(0.85, q1, f'Q1: {q1:.2f}',ha='left',va='center',fontweight='bold')
plt.text(0.85,q3,f'Q3: {q3:.2f}',ha='left',va='center',fontweight='bold')
plt.text(0.85, min_value, f'Min: {min_value:.2f}',ha='left',va='center',fontweight='bold')
plt.text(0.85, max_value, f'Max: {max_value:.2f}',ha='left',va='center',fontweight='bold')

plt.show()
```

# HISTOGRAM FOR DISPLACEMENT ATTRIBUTE

```python
plt.figure(figsize=(8,5))
sns.histplot(df['displacement'],kde=False,bins=10)
plt.title('HISTOGRAM FOR DISPLACEMENT')
plt.xlabel('Engine displacement (cubic inches)')
plt.show()
#kde means kernel density
```

# HISTOGRAM FOR HORSEPOWER ATTRIBUTE

```python
plt.figure(figsize=(8,5))
sns.histplot(df['horsepower'],kde=False,bins=10)
plt.title('HISTOGRAM FOR HORSEPOWER')
plt.xlabel('horsepower')
plt.show()
```

# HISTOGRAM FOR WEIGHT ATTRIBUTE

```python
plt.figure(figsize=(8,5))
sns.histplot(df['weight'],kde=False,bins=10)
plt.title("HISTOGRAM FOR WEIGHT")
plt.xlabel("weight")
plt.show()
```

# HISTOGRAM FOR ACCELERATION ATTRIBUTE

```python
plt.figure(figsize=(8,5))
sns.histplot(df['acceleration'],kde=False,bins=10)
plt.title("HISTOGRAM FOR acceleration")
plt.xlabel("acceleration")
plt.show()
```
